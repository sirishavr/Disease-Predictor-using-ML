Splitting Dataset

```python
# Split the data into training and testing sets

x_train, y_train = train_df.loc[:,train_df.columns != "prognosis"], train_df.loc[:,"prognosis"]
x_test, y_test = test_df.loc[:,train_df.columns != "prognosis"], test_df.loc[:,"prognosis"]
```

Model - Naive Bayes

```python
model = GaussianNB()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

accuracy = metrics.accuracy_score(y_test, y_test)
print(f"Accuracy: {accuracy*100}%")
```

Model - Decision Tree Classifier

```python
dt_model = DecisionTreeClassifier(random_state = 42)
dt_model.fit(x_train, y_train)

y_pred=dt_model.predict(x_test)

accuracy = metrics.accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy*100}%")
```

Model - Random Forest Classifier

```python
rfc_model = RandomForestClassifier(random_state = 42, n_estimators = 100)
rfc_model.fit(x_train, y_train)

rfc_pred=rfc_model.predict(x_test)

accuracy = accuracy_score(y_test, rfc_pred)
print(f"Accuracy: {accuracy*100}%")
```

Model - Gradient Boosting Classifier

```python
gb_model = GradientBoostingClassifier()
gb_model.fit(x_train, y_train)

gb_pred = gb_model.predict(x_test)

accuracy = accuracy_score(y_test, gb_pred)
print(f"Accuracy: {accuracy*100}%")
```

Model - Logistic Regression

```python
# Convert categorical labels to numerical
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y_train)
```

```python
lr_model = LogisticRegression()
lr_model.fit(x_train, y_train)

lr_pred = lr_model.predict(x_test)

accuracy = accuracy_score(y_test, lr_pred)
print(f"Accuracy: {accuracy*100}%")
```

Model - k-Nearest Neighbors(KNN)

```python
knn_model = KNeighborsClassifier()
knn_model.fit(x_train, y_train)

knn_pred = knn_model.predict(x_test)

accuracy = accuracy_score(y_test, knn_pred)
print(f"Accuracy: {accuracy*100}%")
```

```python
# Save the Logistic Regression model and label encoder for later use in the Flask app
joblib.dump(lr_model, 'model.joblib')
joblib.dump(label_encoder, 'label_encoder.joblib')
```
